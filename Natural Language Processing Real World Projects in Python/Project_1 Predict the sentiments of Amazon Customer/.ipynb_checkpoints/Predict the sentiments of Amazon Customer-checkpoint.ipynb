### Libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'Reviews.csv') # r to get rid of all the errors.
```

```python
df.head()
```

```python
df.columns
```

```python
df['Helpful%'] = np.where(df['HelpfulnessNumerator']>0,df['HelpfulnessNumerator']/df['HelpfulnessDenominator'],-1)
```

```python
df.head()
```

```python
df['Helpful%'].unique()
```

```python
df['%upvote'] = pd.cut(df['Helpful%'],bins=[-1,0,0.2,0.4,0.6,0.8,1],labels=['Empty','0-20%','20-40%','40-60%','60-80%','80-100%'])
```

```python
df.head()
```

### Analyse upvotes for different score

```python
df.groupby(['Score','%upvote']).agg('count')
```

```python
df_s = df.groupby(['Score','%upvote']).agg({'Id':'count'}).reset_index().rename(columns={'Id':'Count'})
```

```python
df_s
```

### Create Pivot table & HeatMap of above problem statement for the better 

```python
pivot = df_s.pivot_table(index='%upvote',columns='Score')
pivot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(pivot,annot=True,cmap='YlGnBu')
```
